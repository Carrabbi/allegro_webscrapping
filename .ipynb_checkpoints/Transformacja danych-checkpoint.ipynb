```python
import pandas as pd
import re
```

```python
df = pd.read_csv(
    "../data/interim/job_offers_2021-11-14.csv",
    sep=";",
    encoding="UTF-8",
)
df.fillna("N/A", inplace=True)
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df = df.loc[df['salary_currency'] == "PLN"]
```

```python
df['name'] = df['name'].str.lower()
df['location_city'] = df['location_city'].str.lower()
df.head()
```

```python
df['salary_avg'] = (df['salary_low'] + df['salary_high']) / 2
df.head()
```

```python
df['location_city'].unique()
```

```python
df.loc[df['location_city'] == "warsaw", 'location_city'] = "warszawa"
```

```python
df['location_city'].unique()
```

```python
df['is_senior'] = df['name'].apply(lambda x: bool(re.match(r"senior", x)))
```

```python
df.head(10)
```

```python
df.to_csv(
    "../data/processed/job_offers_2021-11-14.csv",
    sep=";",
    encoding="UTF-8",
    index=False
)
```
