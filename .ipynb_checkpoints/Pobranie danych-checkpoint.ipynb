```python
from time import sleep
from selenium.webdriver import Opera
from bs4 import BeautifulSoup
```

```python
def download_page(job_name, page_number, browser):
    url = f'https://nofluffjobs.com/pl/jobs/?criteria=keyword%3D"{job_name}"&page={page_number}'
    browser.get("view-source:" + url)
    html = browser.find_element_by_tag_name('body').text
    
    return BeautifulSoup(html)
```

```python
def is_empty_page(page):
    headers = page.find_all("h2")
    for header in headers:
        if header.text.strip() == "Nie znaleziono ofert.":
            return True
    return False
```

```python
def get_job_offers(job_name, browser):
    results = []
    page_number = 1
    
    while True:
        page = download_page(job_name, page_number, browser)
        
        if is_empty_page(page):
            break
        
        results.append({'page': page_number, 'job': job_name, 'content': str(page)})
        page_number += 1
        sleep(5)
    
    return results
```

```python
jobs = ["data analyst", "data scientist", "data engineer"]
```

```python
browser = Opera(executable_path="./drivers/operadriver.exe")

data = []
for job_item in jobs:
    data.extend(get_job_offers(job_item, browser))
    
browser.close()
```

```python
data[1]
```

```python
for item in data:
    with open(f"./raw/{item['job']}_{item['page']}.html", 'w',  encoding="utf-8") as out_file:
        out_file.write(item['content'])
            
```
